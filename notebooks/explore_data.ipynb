```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

# Set default Seaborn style
sns.set()

# Set default figure size
sns.set(rc={'figure.figsize':(20, 5)})
```

```python
df = pd.read_csv('data/citi_bike_data_cleaned.csv')
```

# Exploring time variation of trips


```python
df['start_time'] = pd.to_datetime(df['start_time'])

df['start_date'] = df['start_time'].dt.date

df['start_hour'] = df['start_time'].dt.hour

df['start_day'] = df['start_time'].dt.dayofweek

df['start_month'] = df['start_time'].dt.month
```

```python
sns.countplot(x='start_date', data=df)
plt.title('Number of trips by date')
plt.show()
```

```python
sns.countplot(x='start_hour', data=df)
plt.title('Number of trips by hour of the day')
plt.show()
```
